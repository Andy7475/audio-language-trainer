```python
%load_ext autoreload
%autoreload 2

from setup_imports import *  # noqa: F401,F403
from langcodes import Language
from phrases.search import get_phrases_by_collection
from anki_tools import create_anki_deck
from convert import get_collection_title

COLLECTION = "EXTRA"
DECK = "Ordinals"
collection_title = get_collection_title(COLLECTION)
```

# Flash Card Creation

```python
PHRASES = get_phrases_by_collection(COLLECTION, DECK)
print(len(PHRASES))
```

```python

TARGET_LANGUAGE = Language.get("sv-SE")
print(TARGET_LANGUAGE.display_name())
SOURCE_LANGUAGE = Language.get("en-GB")
print(SOURCE_LANGUAGE.display_name())
```

```python
for i, p in enumerate(PHRASES):
    print(f"{i}: {p.english}")
    p.translate(target_language=TARGET_LANGUAGE, refine=True, split_on_space=True)
    p.generate_audio("flashcard", TARGET_LANGUAGE, split_on_space=True)
    p.upload(TARGET_LANGUAGE)
```

# Build several packs

```python
deck_range = range(1, 2)
decks = [f"Pack{i:02d}" for i in deck_range]
print(TARGET_LANGUAGE.display_name())
target_language_tag = TARGET_LANGUAGE.to_tag()
language_name = TARGET_LANGUAGE.language_name()
print(decks)
```

```python
for deck in [DECK]:
    phrases = get_phrases_by_collection(COLLECTION, deck)
    

    # for p in phrases:
    #     p.translate(target_language=TARGET_LANGUAGE, refine=True)
    #     p.generate_audio("flashcard", TARGET_LANGUAGE, split_on_space=True)
    #     #p.generate_image()
    #     p.upload()
        
    phrases = sorted(phrases, key=lambda x: len(x.translations[str(TARGET_LANGUAGE)].text))

    create_anki_deck(
        phrases,
        source_language=SOURCE_LANGUAGE,
        target_language=TARGET_LANGUAGE,
        output_path=f"../outputs/decks/{TARGET_LANGUAGE.to_tag()}/{TARGET_LANGUAGE.to_tag()}-{COLLECTION}-{deck}.apkg",
        deck_name=f"FirePhrase - {TARGET_LANGUAGE.language_name()}::{collection_title}::{deck}",
    )
```
