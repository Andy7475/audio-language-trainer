# Build several packs

```python
decks = ["Pack06", "Pack07", "Pack08", "Pack09", "Pack10", "Pack11", "Pack12", "Pack13", "Pack14", "Pack15"]
print(TARGET_LANGUAGE.display_name())
target_language_tag = TARGET_LANGUAGE.to_tag()
language_name = TARGET_LANGUAGE.language_name()
```

```python
for deck in decks:
    phrases = get_phrases_by_collection(COLLECTION, deck)
    collection_title = get_collection_title(COLLECTION)
    

    for p in phrases:
        p.translate(target_language=TARGET_LANGUAGE, refine=True)
        p.generate_audio("flashcard", TARGET_LANGUAGE, split_on_space=True)
        p.generate_image(TARGET_LANGUAGE)
        p.upload(TARGET_LANGUAGE)
        
    phrases = sorted(phrases, key=lambda x: len(x.translations[str(TARGET_LANGUAGE)].text))
    
    create_anki_deck(
        phrases,
        source_language="en-GB",
        target_language=TARGET_LANGUAGE,
        output_path=f"../outputs/decks/{target_language_tag}/{target_language_tag}-{COLLECTION}-{deck}.apkg",
        deck_name=f"FirePhrase - {language_name}::{collection_title}::{deck}",
    )
```
