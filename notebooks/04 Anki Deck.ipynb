```python
%load_ext autoreload
%autoreload 2

from setup_imports import *  # noqa: F401,F403
from langcodes import Language
from src.phrases.search import get_phrases_by_collection
from random import shuffle
from src.anki_tools import create_anki_deck

COLLECTION = "LM1000"
DECK = "Pack01"
```

# Flash Card Creation

```python
phrases = get_phrases_by_collection(COLLECTION, DECK)
TARGET_LANGUAGE = Language.get("ru-RU")
print(TARGET_LANGUAGE.display_name())
```

```python
for i, p in enumerate(phrases):
    print(f"{i}: {p.english}")
    p.translate(target_language=TARGET_LANGUAGE)
    p.generate_audio("flashcard", TARGET_LANGUAGE, overwrite=True)
    p.upload(TARGET_LANGUAGE, overwrite=True)
```

```python
shuffle(phrases)  # otherwise it's alphabetical
target_language_tag = TARGET_LANGUAGE.to_tag()
language_name = TARGET_LANGUAGE.language_name()
create_anki_deck(
    phrases,
    source_language="en-GB",
    target_language=target_language_tag,
    output_path=f"../outputs/decks/{target_language_tag}/{COLLECTION}-{DECK}.apkg",
    deck_name=f"FirePhrase - {language_name}::{COLLECTION}::{DECK}v2",
)
```
